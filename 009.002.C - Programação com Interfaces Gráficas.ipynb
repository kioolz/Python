```python
# Definindo a classe Sudoku
class Sudoku:
    # O método construtor que inicializa a classe com o tabuleiro do jogo
    # Método para criar uma matriz que preenche o tabuleiro
    def __init__(self, board):
        self.board = board

    # Método para imprimir o tabuleiro do jogo
    def print_board(self):
        for i in range(len(self.board)):
            # Imprime uma linha de separação a cada 3 linhas
            if i % 3 == 0 and i != 0:
                print("- - - - - - - - - - - -")
            for j in range(len(self.board[0])):
                # Imprime um separador a cada 3 colunas
                if j % 3 == 0 and j != 0:
                    print(" | ", end="")
                # Imprime o valor na posição atual
                if j == 8:
                    print(self.board[i][j])
                else:
                    print(str(self.board[i][j]) + " ", end="")

    # Método para encontrar uma célula vazia no tabuleiro
    def find_empty(self):
        for i in range(len(self.board)):
            for j in range(len(self.board[0])):
                # Retorna a posição da célula vazia
                if self.board[i][j] == 0:
                    return (i, j)
        # Retorna None se não houver células vazias
        return None

    # Método para verificar se um número é válido em uma determinada posição
    def valid(self, num, pos):
        # Verifica a linha
        for i in range(len(self.board[0])):
            if self.board[pos[0]][i] == num and pos[1] != i:
                return False

        # Verifica a coluna
        for i in range(len(self.board)):
            if self.board[i][pos[1]] == num and pos[0] != i:
                return False

        # Verifica a caixa
        box_x = pos[1] // 3
        box_y = pos[0] // 3
        for i in range(box_y*3, box_y*3 + 3):
            for j in range(box_x * 3, box_x*3 + 3):
                if self.board[i][j] == num and (i,j) != pos:
                    return False
        # Retorna True se o número é válido
        return True

    # Método para resolver o tabuleiro
    def solve(self):
        find = self.find_empty()
        # Se não houver células vazias, o tabuleiro está resolvido
        if not find:
            return True
        else:
            row, col = find

        # Tenta números de 1 a 9 na célula vazia
        for i in range(1,10):
            if self.valid(i, (row, col)):
                self.board[row][col] = i

                # Recursivamente tenta resolver o tabuleiro com o novo número
                if self.solve():
                    return True

                # Se não puder resolver, desfaz a última tentativa
                self.board[row][col] = 0
        return False


# Cria um tabuleiro de Sudoku
board = [
    [7,8,0,4,0,0,1,2,0],
    [6,0,0,0,7,5,0,0,9],
    [0,0,0,6,0,1,0,7,8],
    [0,0,7,0,4,0,2,6,0],
    [0,0,1,0,5,0,9,3,0],
    [9,0,4,0,6,0,0,0,5],
    [0,7,0,3,0,0,0,1,2],
    [1,2,0,0,0,7,4,0,0],
    [0,4,9,2,0,6,0,0,7]]

# Cria uma instância da classe Sudoku
sudoku = Sudoku(board)

# Imprime o tabuleiro inicial
print("Tabuleiro Inicial:")
sudoku.print_board()

# Resolve o Sudoku
sudoku.solve()

# Imprime o tabuleiro resolvido
print("\nTabuleiro Resolvido:")
sudoku.print_board()
```
