```python
import pandas as pd
from matplotlib import pyplot as plt 

totaldedados = int(input())

TabeladePosicoes=[]
TabeladeTempos=[]

while (len(TabeladePosicoes) < totaldedados) and (len(TabeladeTempos) < totaldedados):
    print("Insira uma posicao")
    X=float(input())
    TabeladePosicoes.append(X)
    print("Insira um tempo")
    T=float(input())
    TabeladeTempos.append(T)
    
ExibeTabela = pd.DataFrame(TabeladePosicoes, TabeladeTempos)
a = ExibeTabela.plot(color = 'red', linestyle='solid')
a.set_title("Trajetória")
a.set_ylabel("Posições")
a.set_xlabel("Tempo")
```
