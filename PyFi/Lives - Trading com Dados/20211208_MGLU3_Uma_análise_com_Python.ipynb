```python
import investpy
import plotly.graph_objects as go
import plotly.subplots
```

```python
#Coletando Preços

mglu = investpy.get_stock_historical_data('MGLU3', country='Brazil', from_date='01/01/2018', to_date='24/11/2021')
```

```python
#Comportamento preços vs IPCA

from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(x=ipca.index, y=ipca['Value'], name='IPCA'), secondary_y=False)
fig.add_trace(go.Scatter(x=mglu.index, y=mglu['Close'], name='MagaluPreço'),secondary_y=True)
fig.update_layout(title_text='Preço vs IPCA')
fig.show()
```

```python
mglu_retornos = mglu['Close'].pct_change()
rol_vol = mglu_retornos.rolling(30).std()
```

```python
#Janela de Voltilidade 30 dias - alta indefinição por parte do mercado

fig = go.Figure()
fig.add_trace(go.Scatter(x=rol_vol.index, y=rol_vol))
fig.update_layout(title_text= 'Janela Volatilidade 30 dias')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=mglu_retornos,name='retornos'))

fig.show()
```

```python
receita_magalu = investpy.get_stock_financial_summary('MGLU3', country='Brazil', summary_type='income_statement', period='quarterly')
receita_magalu.head()
```

```python
precos_mglu = investpy.get_stock_historical_data('MGLU3', country='Brazil', from_date='24/11/2020', to_date='24/11/2021')
precos_mglu
```

```python
#Coletando o numero de ações via Investpy

n_acoes = investpy.get_stock_information('MGLU3', country='Brazil')['Shares Outstanding']
n_acoes = n_acoes
n_acoes
```

```python
import pandas as pd
precos_mglu['PL'] = precos_mglu['Close']/float(((receita_magalu['Net Income']*1000000).sum()/n_acoes))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=precos_mglu.index, y=precos_mglu['PL']))
fig.update_layout(title_text='Preço/Lucro')
fig.show()
```
