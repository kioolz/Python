```python
btcbrl = client.get_klines(symbol = 'BTCBRL', interval = client.KLINE_INTERVAL_1HOUR)
```

```python
type(btcbrl)
```

```python
btcbrl
```

```python
for linha in btcbrl:
  del line[6:]
```

```python
btc_df = pd.DataFrame(btcbrl, columns = ['date', 'open', 'high', 'low', 'close', 'volume', 'close_time', 'asset_volume', 'n of trades', 'asset_column', 'taker_buy', 'based_asset_volume'])
```

```python
btc_df.set_index('date', inplace = True)
```

```python
btc_df.index = pd.to_datetime(btc_df.index, unit = 'ms')
```

```python
btc_df.head()
```

```python
btc_df.shape
```

```python
btc_df.dtypes
```

```python
btc_df['open'] = pd.to_numeric(btc_df['open'])
btc_df['high'] = pd.to_numeric(btc_df['high'])
btc_df['low'] = pd.to_numeric(btc_df['low'])
btc_df['close'] = pd.to_numeric(btc_df['close'])
```

Exercício: Analisar a correlação do Bitcoin com as altcoins. Aparentemente houve mudança no comportamento dessas moedas com relação ao nível de correlação.
